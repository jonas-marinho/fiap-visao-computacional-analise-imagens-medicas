## 4.5 Indicadores de desempenho

O Keras não possui os indicadores de desempenho como precisão, sensibilidade e pontuação f1 por padrão, portanto precisamos implementar externamente.

````python
def recall_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
````

### 4.5.1 Arquitetura ResNet50

**Pergunta**: Explique como é constituída a arquitetura do ResNet50? *Utilize, se necessário, gráficos, projetos que utilizam essa arquitetura. Detalhe também sua topologia em camadas e mostre quais as situações essa arquitetura pode ter mais êxito e quais cenários não tem.*

**Resposta**:

A ResNet50 é uma versão simplificada da rede neural ResNet, utilizando 50 camadas, enquanto a rede completa possui 152 camadas e no ano de 2015 foi o modelo vencedor do desafio de classificação de imagens ImageNet.

Contudo, empilhar o maior número de camadas possíveis em um modelo não é sinônimo de precisão, pois quanto maior o número de camada mais difícil o treinamento se torna e o modelo tende a se degradar.

O problema de agrupar um grande número de camadas de forma sequêncial é que, devido à dinâmica do treinamento por back propagation, os valores propagados para as camadas mais profundas da rede são multiplicados repetidamente, com isso quanto mais profunda a camada maior é a tendência dos valores se saturarem e assim o modelo ser degradado.

A ResNet resolveu esse problema introducindo o conceito Skip Connection:

<p align="center"><img width="400px" src="https://miro.medium.com/max/700/1*kanYOsFl0MmaPk5ZWDjJmw.png"></p>

Com o Skip Connection, existem conexões paralelas às camadas convolucionais, dessa forma as camadas inicias podem se conectar com as camadas que estão a frente na sequência, mas não imediatamente a frente, sem obrigatoriamente passar pelas camadas que estão entre elas, como exemplificado na imagem abaixo:

<p align="center"><img width="600px" src="https://miro.medium.com/max/700/1*uyXEvYztiv3fGGCCPbm8Jg.png"></p>

Com esse tipo de arquitetura, o problema de saturação no back propagation é minimizado, pois são fornecidos caminhos alternativos para os valores de treinamento chegarem até as camadas mais profundas, permitindo que as camadas externas e profundas sejam treinadas de forma equilibrada.

Esse tipo de arquitetura se mostrou muito eficiente para problemas de classificação de imagens, apresentando erro de 3.57% no ImageNet 2015, enquanto o modelo vencedor do ano anterior, o GoogleNet apresentou erro de 6.70%.

__

_Referência: [Understanding and Coding a ResNet in Keras](https://towardsdatascience.com/understanding-and-coding-a-resnet-in-keras-446d7ff84d33)_

A técnica de transfer learning consiste de utilizar o mesmo modelo e treiná-lo para outas imagens. Por tal motivo, excluímos a última camada para modelar com as classes que definimos, ou seja, **controle**, **bacteriana** e **viral**.

Informe a quantidade de classes a serem classificadas.

````python
## IMPLEMENTE

qtde_classes = 3
````

````python
conv_base = ResNet50(include_top=False)

for layer in conv_base.layers:
    layer.trainable = False
````

````python
x = conv_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x) 

predictions = layers.Dense(qtde_classes, activation='softmax')(x)
model = Model(conv_base.input, predictions)

model.summary()
````

````python
optimizer = optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=[precision_score, recall_score, f1_score])
````

O número de épocas define quantas vezes o modelo irá treinar e validar o erro, assim ajustando os pesos para melhor convergência.
Escolha o número adequado de épocas para alcançarmos pelo menos **70% de precisão de validação**.

````python
## IMPLEMENTE

qtde_epocas = 50
````

````python
history = model.fit(train_generator, epochs=qtde_epocas, validation_steps=5, steps_per_epoch=5, 
                    validation_data=val_generator)
````

Um modelo que converge bem possui o gráfico de perda (*loss*) descendente e os gráfico de precisão (*precision*), sensibilidade (*recall*) e pontuação f1 (*f1 score*) em acendente.

````python
# Exibindo dados de Precisão
plt.plot(history.history['precision_score'])
plt.plot(history.history['val_precision_score'])
plt.title('model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Sensibilidade
plt.plot(history.history['recall_score'])
plt.plot(history.history['val_recall_score'])
plt.title('model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de F1 Score
plt.plot(history.history['f1_score'])
plt.plot(history.history['val_f1_score'])
plt.title('model f1_score')
plt.ylabel('f1_score')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Perda
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
````

### 4.5.2 Arquitetura VGG16

**Pergunta**: Explique como é constituída a arquitetura do VGG16? *Utilize, se necessário, gráficos, projetos que utilizam essa arquitetura. Detalhe também sua topologia em camadas e mostre quais as situações essa arquitetura pode ter mais êxito e quais cenários não tem.*

**Resposta**:

A rede neural VGG16 tem o seu funcionamento baseado em camadas convolucionais intercaladas com camadas max pooling, possuindo no final também camadas totalmente conectadas para realizar classificações através da função de ativação softmax. A arquitetura está representada na imagem abaixo:

<p align="center"><img width="600px" src="https://neurohive.io/wp-content/uploads/2018/11/vgg16-1-e1542731207177.png"></p>

As camadas convolucionais utilizam janelas de convolução (strides) do menor tamanho possível no qual ainda seja mantida a noção de direção da posição (cima, baixo, esquerda, direita e centro), sendo este o tamanho de 3x3 pixels. Também utilizando padding nas bordas da imagem, para que dessa forma as dimensões de entrada sejam mantidas na saídas.

Já as camadas de max pooling utilizam strides de 2x2 pixels sem padding, dessa forma as dimensões de horizontais e verticas são reduzidas à metada após passar por estas camadas.

Após o sinal da imagem passar pelas camadas convolucionais e de max pooling, chega às camadas densas, que o classificam através da função de ativição softmax. Em seu projeto original, a VGG16 é capaz de classificar em 1000 classes distintas.

Essa rede neural foi a vencedora do desafio ImageNet de 2014, sua principal diferença em relação à AlexNet, vencedora do desafio do ano anterior, é o tamanho dos strides nas camadas convolucinais, adotando quadros de 3x3 pixels em todas as camadas de convolução, enquanto a AlexNet utilizava quadros de 11x11 e 5x5 pixels nas duas camadas convolucionais iniciais.

O objetivo da rede é classificar imagens, com capacidade para predizer 1000 rótulos distintos, contudo para cenários em que há mais de 1000 rótulos possíveis o desempenho não será tão bom, porém esse problema pode ser resolvido adicionando mais neurônios à última camada densa.

Para classificação de frame de vídeos em tempo real, essa rede não é adequada, pois o tempo que leva para realizar a classificação pode não ser suficiente para uma aplicação desse tipo, sendo mais recomendadas outras tecnologias, como Yolo.

__

_Referência: [VGG16 – Convolutional Network for Classification and Detection](https://neurohive.io/en/popular-networks/vgg16/)_

## 4.8 Abordagem adicional (obrigatório)

Considerando os outros classificadores, escolha outro que ainda não foi utilizado, implemente abaixo. Ao final compare os resultados e explique os resultados.

_Não se esquece de utilizar as importações adequadas para cada modelo.
A forma de implementação deve respeitar as mesmas condições como valor de split e quantidade de imagens para poder comparar os modelos._

````python
#IMPLEMENTE

# Vamos utilizar um modelo baseado na VGG16, porém mais simples com menos camadas de convolução e max pooling. Para tentar superar os resultados obtido 
# com o modelo VGG16, vamos aplicar camadas de tratamento de imagens antes das primeira camada convolucional.

# Criando uma função para aplicar a detecção de bordas
import cv2
def border_detection(img_path):
  img = cv2.imread(img_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img = cv2.resize(img, (224,224))
  img = np.uint8(img)
  img = cv2.Canny(img, 50, 80)
  img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
  return img

# Agora precisamos transformar as imagens de teste e treino originais em imagens com as bordas detectadas
transformed_train_generator = []
img_train_label = []
for img_path, img_label in zip(train_generator.filepaths, train_generator.labels):
  transformed_train_generator += [border_detection(img_path)]
  img_array_label = [0,0,0]
  img_array_label[img_label] = 1
  img_train_label += [img_array_label]
transformed_train_generator = np.array(transformed_train_generator, dtype=np.float64)
img_train_label = np.array(img_train_label, dtype=np.uint8)

transformed_val_generator = []
img_val_label = []
for img_path, img_label in zip(val_generator.filepaths, val_generator.labels):
  transformed_val_generator += [border_detection(img_path)]
  img_array_label = [0,0,0]
  img_array_label[img_label] = 1
  img_val_label += [img_array_label]
transformed_val_generator = np.array(transformed_val_generator, dtype=np.float64)
img_val_label = np.array(img_val_label, dtype=np.uint8)

model = Sequential()
model.add(Conv2D(64, (3,3), padding='same'))
model.add(Conv2D(64, (3,3), padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128, (3,3), padding='same'))
model.add(Conv2D(128, (3,3), padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(256, (3,3), padding='same'))
model.add(Conv2D(256, (3,3), padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(layers.GlobalAveragePooling2D())
model.add(Dense(128))
model.add(Dense(3, 'softmax'))

model.build(input_shape=(None,224,224,3))
model.summary()
````

````python
optimizer = optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=[precision_score, recall_score, f1_score])
````

````python
history = model.fit(x=transformed_train_generator, y=img_train_label, epochs=qtde_epocas, steps_per_epoch=5)
````

````python
# Testando utilizando os dados de teste
model.evaluate(x=transformed_val_generator, y=img_val_label, batch_size=60)
````
